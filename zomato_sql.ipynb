```python
import pandas as pd 
import sqlite3 as sql
```

```python
df = pd.read_csv('zomato.csv')
```

```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("zomato", con =engine)
```

```python
## Test the code by select the complete table

sql='''

Select * from zomato
limit 5


''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Count of Distinct Vendors

sql='''

Select 'The total number of vendors :- '||count(Distinct url) as Value from zomato

''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Areas which with top vendors (for immediate food access and availability)

sql='''

Select location,count(Distinct url) as Value from zomato
group by 1
order by count(Distinct url) desc
''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## count of restaurants that have performed more than the average rating. Select a breakdown areawise

## Areas with vendos which have top ratings


sql='''

with b as (
select 
location,
name,
rate,
votes
from zomato
group by 1,2,3,4)

select * from
(select
s.location,
sum(case when s.rating>=3.7 then 1 else 0 end) as Above_avg_restuarants_count
from
(select
b.location,
b.name,
b.rate,
round(avg(b.rate),2) as rating,
b.votes
from b
where location!='None' and rate !='None'
group by 1,2,3,5 having b.votes >= avg(b.votes))s
group by 1) a
order by a.Above_avg_restuarants_count desc
''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Area wise performance (TOtal number of vendors greater than avg rating and overall avg rating of the food)
sql='''

with b as (
select 
location,
name,
rate,
votes
from zomato
group by 1,2,3,4),

d as (
select location,
round(avg(rate),2) as rating from zomato
group by 1
order by round(avg(rate),2) desc)

select
a.location,
a.Above_avg_restuarants_count,
d.rating
from
(select
s.location,
sum(case when s.rating>=3.7 then 1 else 0 end) as Above_avg_restuarants_count
from
(select
b.location,
b.name,
b.rate,
round(avg(b.rate),2) as rating,
b.votes
from b
where location!='None' and rate !='None'
group by 1,2,3,5 having b.votes >= avg(b.votes))s
group by 1) a
join d on a.location=d.location
order by a.Above_avg_restuarants_count desc
''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```
